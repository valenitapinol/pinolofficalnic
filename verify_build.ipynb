# Verificación de entorno y build
Este cuaderno ejecuta comprobaciones básicas del entorno de Node.js y Gradle para el proyecto PinolApp.

```python
import subprocess
import json
import os
from pathlib import Path

repo_root = Path('/workspaces/pinolofficalnic')
assert repo_root.exists(), 'No se encontró el repositorio en /workspaces/pinolofficalnic'
```

```python
def run_command(cmd, cwd=None, capture_output=True, check=False):
    result = subprocess.run(cmd, shell=True, cwd=cwd, capture_output=capture_output, text=True)
    return {
        'cmd': cmd,
        'cwd': cwd,
        'returncode': result.returncode,
        'stdout': result.stdout,
        'stderr': result.stderr,
    }

def check_environment():
    commands = [
        'node --version',
        'npm --version',
        'java -version',
        'gradle --version'
    ]
    results = []
    for cmd in commands:
        results.append(run_command(cmd, cwd=repo_root))
    return results
```

```python
results = check_environment()
for res in results:
    print('CMD:', res['cmd'])
    print('RETURN:', res['returncode'])
    print(res['stdout'])
    print(res['stderr'])
    print('---')
```
